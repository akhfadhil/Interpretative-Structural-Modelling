```python
import numpy as np
import pandas as pd
```

```python
# Inisiasi matriks numpy
ordo = int(input("Enter the number of ordo: "))

matrix = np.diag(np.full(ordo,'X'))     
```

```python
# Mengisi segitiga atas matrix
for i in range(ordo):
    for j in range(ordo):
        if i != j and i<j:  # Skip elements with the same row and column index
            matrix[i][j] = input("Enter value "+ str(i+1)+ "&" + str(j+1) + ": ")
```

```python
# Mengisi segitiga bawah matrix otomatis
for i in range(ordo):
    for j in range(ordo):
        if i != j and i>j:  # Skip elements with the same row and column index
            if matrix[j][i] == 'V':
                 matrix[i][j] = 'N'
            elif matrix[j][i] == 'A':
                 matrix[i][j] = 'M'
            elif matrix[j][i] == 'X':
                 matrix[i][j] = 'X'
            elif matrix[j][i] == 'O':
                 matrix[i][j] = 'O'
```

```python
# Mengubah ke biner
for i in range(ordo):
    for j in range(ordo):
        if matrix[i][j] == 'V':
            matrix[i][j] = 1
        elif matrix[i][j] == 'A':
            matrix[i][j] = 0
        elif matrix[i][j] == 'X':
            matrix[i][j] = 1
        elif matrix[i][j] == 'O':
            matrix[i][j] = 0
        elif matrix[i][j] == 'M':
            matrix[i][j] = 1
        elif matrix[i][j] == 'N':
            matrix[i][j] = 0
```

```python
matrix
```

```python
# Mengubah matrix numpy ke dataframe
dataset = pd.DataFrame(matrix)
dataset
```

```python
# Membuat kolom drive power
dataset["Dr P"] = dataset.apply(lambda row: sum(row[0:ordo]=='1') ,axis=1)
```

```python
# Membuat kolom dependence power
lst = []
for column in range(len(dataset)):
    DeP = dataset[column].value_counts()['1']
    lst.append(DeP)

dataset['De P'] = lst
```

```python
dataset
```

```python
# Menentukan titik potong (rata-rata drive power/dependence power)
dataset["Dr P"].mean()
dataset["De P"].mean()

# Seleksi kuadran 3 atau linkage
for row in range(len(dataset)):
    if dataset['Dr P'][row] > dataset["Dr P"].mean() and dataset['De P'][row] > dataset["De P"].mean():
        print(str(row) + " " + str(dataset['Dr P'][row]) + " " +str(dataset['De P'][row]))
```
